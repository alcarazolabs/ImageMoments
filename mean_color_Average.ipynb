```python
#Author: Freddy Alcarazo
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('limon.jpg')
plt.imshow(image)
plt.show()
```

```python
r_channel = image.copy()
```

```python
r_channel[:, :, 0] = 0 #remove blue channel
r_channel[:, :, 1] = 0 #remove gree channel
```

```python
plt.imshow(r_channel)
plt.show()
```

```python
r_array = np.asarray(r_channel)
```

```python
r_array.shape
```

```python
y = np.ma.masked_where(r_array == 0, r_array)
```
